```python
import pandas as pd
from datetime import datetime

estudiantes={
    'Nombre':['Juan','Ana','Carlos','Maria','Luis'],
    'Edad':[18,22,20,19,21],
    'Ciudad':['Quito','Guayaquil','Loja','Cuenca','Ambato'],
    'Promedio':[9.5,8.0,7.5,9.0,8.5]
}
dt=pd.DataFrame(estudiantes)
print(dt)

print("\n Promedio \n")
print(dt['Promedio'])

dt['Aprobado']=dt['Promedio']>8
print("\n Aprobado\n")
print(dt)

aprobado=dt[dt['Promedio']>8]
print("\n Estudiantes Aprobados\n")
print(aprobado)

estudiantes_aprobados=aprobado['Nombre']
print("\n Estudiantes Aprobados\n")
print(estudiantes_aprobados)

eventos={
    'Evento': ['Concierto', 'Conferencia', 'Seminario', 'Feria', 'Taller'],
    'Fecha': ['2025-05-01', '2025-06-15', '2025-07-10', '2025-08-25', '2025-09-05'],
    'Lugar': ['Quito', 'Guayaquil', 'Loja', 'Cuenca', 'Ambato']

}
dt=pd.DataFrame(eventos)
fechas=pd.to_datetime(eventos['Fecha'])
print("\n Fechas\n")
print(fechas)

dt['Fechas']=fechas
dt['Anio']=dt['Fechas'].dt.year
dt['Mes']=dt['Fechas'].dt.month
print("\n Convercion Fechas\n")
print(dt)

dt['Fechas']=fechas
dt['Anio']=dt['Fechas'].dt.year
dt['Mes']=dt['Fechas'].dt.month
print("\n Convercion Fechas\n")
print(dt)

solo_2025=dt[dt['Anio']==2025]
print("\n Solo 2020\n")
print(solo_2025[['Evento','Lugar']])

solo_evento=dt[dt['Mes']==9]
print("\n Solo Evento\n")
print(solo_evento['Evento'])

fecha_actual=datetime.now()
dt['Días'] = dt['Fechas'].apply(lambda x: (x - fecha_actual).days)
print("\n Dias\n")
print(dt)

productos={
    'Producto': ['Laptop', 'Celular', 'Tablet', 'Auriculares', 'Teclado'],
    'Precio': [500, 300, 150, 50, 75],
    'Cantidad Vendida': [10, 15, 7, 25, 30],
    'Fecha de Venta': ['2025-04-01', '2025-04-03', '2025-04-05', '2025-04-06', '2025-04-10']
}
dt=pd.DataFrame(productos)
dt['Fecha de Venta']=pd.to_datetime(dt['Fecha de Venta'])

dt['Ingreso Total']=dt['Precio']*dt['Cantidad Vendida']
print("\n Ganancia\n")
print(dt)

ingreso_mayor=dt[dt['Ingreso Total']>1000]
print("\n Ganancia Mayor\n")
print(ingreso_mayor)

dt['Mes']=dt['Fecha de Venta'].dt.month
print("\n Aniadir Mes\n")
print(dt)

ventas_abril=dt[dt['Mes']==4]
print("\n Solo Abril\n")
print(ventas_abril[['Mes','Ingreso Total']])

temperaturas=pd.Series([18.5, 20.3, 22.0, 21.8, 19.7, 20.1, 21.4],
                       index=['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo'])
print("\n Temperaturas\n")
seleccion=temperaturas.loc[['Lunes','Miercoles','Viernes']]
print(seleccion)

temperatura_media=round(temperaturas.mean(),3)
print("\n Temperatura Media\n")
print(temperatura_media,"Grados Centigrados")

temperatura_farenheit=round(temperaturas*(9/5)+32,2)
print("\n Temperatura Farenheit\n")
print(temperatura_farenheit)

temperatura_alta=temperaturas.max()
temperatura_baja=temperaturas.min()
dia_temperatura_alta = temperaturas.idxmax()
dia_temperatura_baja = temperaturas.idxmin()
print("\n Maximo y minimo temperaturas\n")
print("Día de la temperatura más alta:", dia_temperatura_alta,"con: ",temperatura_alta)
print("Día de la temperatura más baja:", dia_temperatura_baja,"con: ",temperatura_baja)

asistencia={
      'Estudiante': ['Ana', 'Luis', 'Carlos', 'Maria', 'Pedro'],
      'Días Asistidos': [20, 18, 22, 15, 19],
      'Mes': ['Abril', 'Abril', 'Abril', 'Abril', 'Abril']
}
dt=pd.DataFrame(asistencia)
dt['Total de Días en el Mes']=30
print(dt)

porcentaje_asistencia=round((dt['Días Asistidos']/dt['Total de Días en el Mes'])*100,2)
dt['Porcentaje de Asistencia']=porcentaje_asistencia
print(dt)

mas_asistencia=dt[dt['Porcentaje de Asistencia']>80]
print("\n Mas Asistencia\n")
print(mas_asistencia['Estudiante'])

ordenar_asistencia=dt.sort_values(by='Porcentaje de Asistencia',ascending=False)
print("\n Ordenar Asistencia\n")
print(ordenar_asistencia[['Estudiante','Porcentaje de Asistencia']])

empresa={
      'Fecha': ['2025-03-01', '2025-03-05', '2025-03-10', '2025-03-15', '2025-03-20'],
      'Ingreso': [1500, 1200, 2000, 1800, 1600],
      'Egreso': [500, 700, 600, 800, 400]
}
dt=pd.DataFrame(empresa)
dt['Fecha']=pd.to_datetime(dt['Fecha'])

dt['Saldo']=dt['Ingreso']-dt['Egreso']
print(dt)

totales=dt[['Ingreso','Egreso','Saldo']].astype(float).sum()
print("\n Totales\n")
print(totales)

saldo_positivo=dt[dt['Saldo']>0]
print("\n Saldo Positivo\n")
print(saldo_positivo[['Fecha','Saldo']])
```
