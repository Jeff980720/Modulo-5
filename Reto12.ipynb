```python
import numpy as np
import pandas as pd
columnas=["Cantidad Vendida","Precio Unitario","Descuento %","Costo de Produccion"]
matriz_ventas=np.array([
    [30,50,10,20], ##Prducto 1
    [20,70,5,30], ##Prducto 2
    [50,20,15,10], ##Prducto 3
    [15,100,20,40], ##Prducto 4
    [10,150,25,60], ##Prducto 5
    [25,200,10,80], ##Prducto 6
    [40,30,0,15], ##Prducto 7
    [35,120,5,50], ##Prducto 8
    [45,60,20,25], ##Prducto 9
    [50,90,30,35], ##Prducto 10
    [60,110,15,45], ##Prducto 11
    [30,80,10,30], ##Prducto 12
    [25,50,5,20], ##Prducto 13
    [40,150,0,60], ##Prducto 14
    [55,60,10,25], ##Prducto 15
    [35,70,10,35], ##Prducto 16
    [45,100,20,45], ##Prducto 17
    [50,120,15,50], ##Prducto 18
    [60,80,10,40], ##Prducto 19
    [30,150,25,55], ##Prducto 20
])
df_ventas=pd.DataFrame(matriz_ventas,columns=columnas)
print(df_ventas)
# df_ventas.to_excel("Ventas.xlsx")
# Ejercicio 1
print("Ejercicio 1")
cantidad_vendida=matriz_ventas[:,0]
print("Cantidad Vendida: ",cantidad_vendida)
precio_unitario=matriz_ventas[:,1]
print("Precio Unitario: ",precio_unitario)
descuento=matriz_ventas[:,2]
print("Descuento: ",descuento)
costo_produccion=matriz_ventas[:,3]
print("Costo de Produccion: ",costo_produccion)
# Ejercicio 2
print("Ejercicio 2")
total_ingreso=cantidad_vendida*precio_unitario
ingreso_descuento=total_ingreso*(1-descuento/100)
for i, ingreso in enumerate(ingreso_descuento):
    print(f"Producto {i+1}: Ingreso despues del descuento = {ingreso}")
# Ejercicio 3
print("Ejercicio 3")
total_global=sum(ingreso_descuento)
print("Total Global de los 20 productos: ",total_global)
# Ejercicio 4
print("Ejercicio 4")
costo_total=cantidad_vendida*costo_produccion
for i, costo in enumerate(costo_total):
    print(f"Producto {i+1}. Su costo total es de:  = {costo}")
# Ejercicio 5
print("Ejercicio 5")
ganancia_bruta=ingreso_descuento-costo_total
for i, ganancia in enumerate(ganancia_bruta):
     print(f"Producto {i+1}. Su ganancia bruta es de:  = {ganancia}")
# Ejercicio 6
print("Ejercicio 6")
total_global_bruta=sum(ganancia_bruta)
print("Total Global de las ganancias bruta: ",total_global_bruta)
# Ejercicio 7
print("Ejercicio 7")
ingreso_total_promedio=np.mean(ingreso_descuento)
print("Ingreso Total Promedio: ",ingreso_total_promedio)
# Ejercicio 8
print("Ejercicio 8")
precio_promedio=np.mean(precio_unitario)
print("Precio Promedio: ",precio_promedio)
# Ejercicio 9
print("Ejercicio 9")
producto_rentable=np.max(ingreso_descuento)
indice=np.argmax(ingreso_descuento)
print(f"El Productos mas rentable es el Producto {indice+1} con: {producto_rentable}")
# Ejercicio 10
print("Ejercicio 10")
mayor_ganancia_bruta=np.max(ganancia_bruta)
indice=np.argmax(ganancia_bruta)
print(f"El Productos con mayor ganancia bruta es el Producto {indice+1} con: {mayor_ganancia_bruta}")
# Ejercicio 11
print("Ejercicio 11")
total_cantidad_vendida=np.sum(cantidad_vendida)
print("Total de Cantidad Vendida: ",total_cantidad_vendida)
# Ejercicio 12
print("Ejercicio 12")
promedio_descuento=np.mean(descuento)
print("Promedio de Descuento: ",promedio_descuento,"%")
# Ejercicio 13
print("Ejercicio 13")
costo_total_produccion=np.sum(costo_produccion)
print("Costo Total de Produccion: ",costo_total_produccion)
# Ejercicio 14
print("Ejercicio 14")
porcentaje_ingreso_total=np.round(total_global/np.sum(costo_produccion),2)
print("El Porcentaje de Ingreso Total ante el costo es de: ",porcentaje_ingreso_total,"%")
# Ejercicio 15
print("Ejercicio 15")
rentabilidad=np.round(total_global_bruta/total_global,2)
print("La rentabilidad es de: ",rentabilidad,"%")
```
